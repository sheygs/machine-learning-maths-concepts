## Modelling Systems of Linear Equations as Matrices

```python
# import required packages
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_lines
```

```python
A = np.array([
   [-1, 3],
   [3, 2]
], dtype=float)

b = np.array([7, 1], dtype=np.dtype(float))

print('Matrix A: \n', A)
print('Output b: \n', b)
```

```python
# determine the dimensions of the matrices
print(f'Shape of A: {A.shape}')
print(f'Shape of b: {b.shape}')
```

```python
# solve the linear algebra
values = np.linalg.solve(A, b)
print(f'solution: {values}')
```

```python
# compute the determinant of the matrix A
d = np.linalg.det(A)

# format to 1.d.p
print(f'determinant: {d: .1f}') # unique solution
```

### 2 - Visualising a 2x2 systems as Plotlines

```python
A_system = np.hstack((A, b.reshape(2, 1)))
print(f'A_system: {A_system}')
print(f'second row matrix: {A_system[1]}')

# plot the graph solution
plot_lines(A_system)
```

### 3 - Systems of Linear Equations with no Solutions

Let's find the determinant of the corresponding matrix for another system of linear equations:

$$
\begin{cases} 
-x_1+3x_2=7, \\ 3x_1-9x_2=1, \end{cases}\tag{2}
$$

```python
A_2 = np.array(
   [
      [-1, 3],
      [3, -9]
   ], dtype=float)

b_2 = np.array([7, 1], dtype=float)

d_2 = np.linalg.det(A_2)
print(f'determinant of matrix d_2: {d_2: .2f}')
```

The determinant is 0 (zero) meaning that the system has no unique solution. Solving the linear systems of equations will give an error due to singularity

```python
try:
    values = np.linalg.solve(A_2, b_2)
    print(f'values: {values}')
except np.linalg.LinAlgError as err:
    print(f'error : {err}')
```

```python
A_2_system = np.hstack((A_2, b_2.reshape(2, 1)))
print(f'A_2_system: {A_2_system}')

# plot the graph
# two lines of equation are parallel
plot_lines(A_2_system)
```

### 4 - Systems of Linear Equations with Infinite Number of Solutions


By changing the free coefficients of the system $(2)$, you can bring it to consistency:

$$
\begin{cases} 
-x_1+3x_2=7, \\ 3x_1-9x_2=-21, \end{cases}\tag{3}
$$

```python
b_3 = np.array([-7, 21], dtype=float)

A_3_system = np.hstack((A_2, b_3.reshape((2, 1))))
print(f'A_3_system: {A_3_system}')
```

```python
# plot the graph
# infinite solutions
plot_lines(A_3_system)
```
